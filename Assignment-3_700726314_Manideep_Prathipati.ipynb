```python
import numpy as np

# Generate random vector of size 15 with integers in range 1-20
random_vector = np.random.randint(low=1, high=21, size=15)

# Reshape the vector to a 3 by 5 matrix
matrix_3x5 = random_vector.reshape(3, 5)

print(matrix_3x5)

print()

# Find the index of the maximum value in each row
max_indices = np.argmax(matrix_3x5, axis=1)

# Replace the maximum value in each row with 0
for i, j in enumerate(max_indices):
    matrix_3x5[i, j] = 0

print(matrix_3x5)
```

```python
import numpy as np

# Define the square array
array = np.array([[3, -2], [1, 0]])

print("Printing the Original square array:\n",array)

print()

# Compute the eigenvalues and right eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(array)

# Print the eigenvalues and eigenvectors
print("Eigenvalues:", eigenvalues)
print("Eigenvectors: \n", eigenvectors)
```

```python
import numpy as np

# Define the array
array = np.array([[0, 1, 2], [3, 4, 5]])

print("Printing the Original square array:\n",array)

print()

# Compute the sum of the diagonal elements
sum_diagonal = np.trace(array)

# Print the sum of the diagonal elements
print("Sum of diagonal elements:", sum_diagonal)
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
sizes = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
explode = (0.1, 0, 0, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Java')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')

plt.show()
```
